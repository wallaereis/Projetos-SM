```python
import pandas as pd
import matplotlib.pyplot  as plt
import seaborn as sns   

df = pd.read_excel(r"P:\Indicadores\Diagnóstico\Premiações\Recibos de Premiação\2025\Premiação - Recibos - Set25\Recibo rascunho - Setembro - 2025.xlsx")
```

```python
df = df[['Natureza RH', 'Natureza', 'SETOR', 'REFERÊNCIA QUE GEROU O PAGAMENTO',
       'AUTORIZADOR', 'RESPONSÁVEL EM TRAZER O RECIBO', 'NOME DO RECEBEDOR',
       'LOJA', 'VALOR CASH', 'VALOR $ RECEBIDO  WIIPO', 'CPF']]
```

```python
df  = df[~df['SETOR'].isnull()]
```

```python
df.loc[df['LOJA']=='Lote XV', 'LOJA'] = 'Lote Xv'
```

```python
df1 = df[~((df['SETOR'].str.contains('coord',case=False, na=False)) | (df['SETOR'].str.contains('gerencia',case=False, na=False)) | (df['SETOR'].str.contains('regional',case=False, na=False))) ]
```

```python
def filtro_lojas(df1, loja_nome):
    colunas = ['Natureza RH', 'Natureza', 'SETOR', 'REFERÊNCIA QUE GEROU O PAGAMENTO',
               'AUTORIZADOR', 'RESPONSÁVEL EM TRAZER O RECIBO', 'NOME DO RECEBEDOR',
               'LOJA', 'VALOR CASH', 'VALOR $ RECEBIDO  WIIPO', 'CPF']
    return df1[df1['LOJA'] == loja_nome][colunas]

```

```python
fonseca = filtro_lojas(df1, 'Fonseca')
piratininga = filtro_lojas(df1, 'Piratininga')
itaipu = filtro_lojas(df1, 'Itaipú')
santa_cruz = filtro_lojas(df1, 'Santa Cruz')
lote_xv = filtro_lojas(df1, 'Lote Xv')
porto_novo = filtro_lojas(df1, 'Porto Novo')
batalha = filtro_lojas(df1, 'Batalha')
nv_iguacu = filtro_lojas(df1, 'Nova Iguaçú')
porto_velho = filtro_lojas(df1, 'Porto Velho')
engenho = filtro_lojas(df1, 'Engenho')
mesquita = filtro_lojas(df1, 'Mesquita')
valverde = filtro_lojas(df1, 'Valverde')
barreto = filtro_lojas(df1, 'Barreto')
nv_america = filtro_lojas(df1, 'Nova América')
inhauma = filtro_lojas(df1, 'Inhaúma')
curtume = filtro_lojas(df1, 'Curtume')
```

```python
with pd.ExcelWriter(r"C:\Users\wallace.souza\Downloads\Funcionarios.xlsx", engine='openpyxl', mode='w') as writer:
    fonseca.to_excel(writer, sheet_name='Fonseca', index=False)
    piratininga.to_excel(writer, sheet_name='Piratininga', index=False)
    itaipu.to_excel(writer, sheet_name='Itaipú', index=False)
    santa_cruz.to_excel(writer, sheet_name='Santa Cruz', index=False)
    lote_xv.to_excel(writer, sheet_name='Lote Xv', index=False)
    porto_novo.to_excel(writer, sheet_name='Porto Novo', index=False)
    batalha.to_excel(writer, sheet_name='Batalha', index=False)
    nv_iguacu.to_excel(writer, sheet_name='Nova Iguaçú', index=False)
    porto_velho.to_excel(writer, sheet_name='Porto Velho', index=False)
    engenho.to_excel(writer, sheet_name='Engenho', index=False)
    mesquita.to_excel(writer, sheet_name='Mesquita', index=False)
    valverde.to_excel(writer, sheet_name='Valverde', index=False)
    barreto.to_excel(writer, sheet_name='Barreto', index=False)
    nv_america.to_excel(writer, sheet_name='Nova América', index=False)
    inhauma.to_excel(writer, sheet_name='Inhaúma', index=False)
    curtume.to_excel(writer, sheet_name='Curtume', index=False)
    
    
 
```
