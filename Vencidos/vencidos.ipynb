```python
import pandas as pd

df = pd.read_excel(r"C:\Users\wallace.souza\Desktop\Scripts\Vencidos\Vencidos\Vencidos de Outubro 2025.xlsx")
cd = pd.read_excel(r"C:\Users\wallace.souza\Desktop\Scripts\6 dias\004075299_006119963_20250908_071714_cadastro_consinco.xlsx")
to = pd.read_excel(r"C:\Users\wallace.souza\Desktop\Scripts\Vencidos\Vencidos\Pasta1.xlsx")
```

```python
df = df[(df["Data da Auditoria:"].dt.month  == 10) & (df["Data da Auditoria:"].dt.year  == 2025)]
```

```python
df1 = df.merge(to[["x.1","x.4","x.13","x.3"]], left_on="EAN do Produto:", right_on="x.4", how='left')
```

```python
df1 = df1.merge(cd[['SEQPRODUTO','SETOR']], left_on="x.1", right_on="SEQPRODUTO", how="left")
```

```python
df1 = df1.drop_duplicates(["Local do Solicitante:","QTD do Produto:",'SEQPRODUTO',"Dias de Vencido","Local onde o produto foi encontrado:","Hora de conclusão"])
```

```python
df1.loc[df1['SETOR'] == 'ESPECIARIAS', 'SETOR'] = 'FLV'
```

```python
df1 = df1.sort_values(['SETOR','SEQLOJA'])
```

```python
df1 = df1[['Hora de conclusão', 'Local do Solicitante:', 'Colaborador:',
       'Local onde o produto foi encontrado:', 'EAN do Produto:',
       'QTD do Produto:', 'Data da Auditoria:',
       'Vencido em:', 'Dias de Vencido', 'SEQPRODUTO','x.3', 'x.13','SETOR']]#.to_excel(r"C:\Users\wallace.souza\Downloads\2.xlsx", index=False)
```

```python
df1 = df1.groupby("Local do Solicitante:")['SETOR'].value_counts().reset_index()
```

```python
df1['TOTAL'] = df1.groupby("Local do Solicitante:")['count'].transform('sum')
```
