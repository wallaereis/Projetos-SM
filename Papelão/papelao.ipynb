```python
import pandas as pd

df = pd.read_excel(r"C:\Users\wallace.souza\Downloads\Medição Outubro 2025.xlsx")
df1 = pd.read_excel(r"C:\Users\wallace.souza\Downloads\Consolidado de Reciclagem - 2025 2.0.xlsx")
```

```python
df.columns = ['Data',	'Material',	'Nº Ticket'	,'Peso CRR'	,'R$/Unit'	,'R$ CRR']
df = df.drop(index=[0,1,2], axis=0)
df = df.dropna(subset='R$ CRR')
df['Data'] = pd.to_datetime(df['Data'], format='%d-%m-%Y')
df['Data'] = df['Data'].dt.strftime('%d/%m/%Y')
df = df.fillna("")
```

```python
df1.columns = df1.loc[0]
df1 = df1.drop(index=0, axis=0)
```

```python
df1.columns = df1.columns.str.strip()
```

```python
df1 = df1[[ 'PRODUTO','DATA','DATA FIXA','PESAGEM KG']]
```

```python
df1.loc[df1["PRODUTO"] == 'PAPELÃO', 'PRODUTO'] = 'Papelão'
df1.loc[df1["PRODUTO"] == "Plástico transparente", 'PRODUTO'] = 'Cristal'
```

```python
df1["DATA FIXA"]= pd.to_datetime(df1["DATA FIXA"])
df1["DATA"]= pd.to_datetime(df1["DATA"])
df["Data"] = pd.to_datetime(df["Data"], dayfirst=True, errors="coerce")
```

```python
df1 = df1[df1["DATA FIXA"] == "2025-10-01"]
```

```python
df1.columns
```

```python
df = df.merge(df1, left_on=['Peso CRR','Material','Data'], right_on=['PESAGEM KG','PRODUTO','DATA'], how='outer')
```

```python
df.drop_duplicates("Nº Ticket").to_excel(r"C:\Users\wallace.souza\Downloads\papelao.xlsx")
```
